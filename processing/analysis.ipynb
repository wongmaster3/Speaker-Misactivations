```python
import pandas as pd
import math
import numpy as np
```

```python
light_activation_filename = "echo_com_2_light_activations.csv"
word_list_filename = "nest_word_generations.csv"
network_filename = "tmmp_network.csv" # This is the file generated by analysis.py
```

```python
light_activations = pd.read_csv(light_activation_filename)
```

```python
words = pd.read_csv(word_list_filename)
```

```python
pd.set_option('display.float_format', lambda x: '%ld' % x)
```

```python
network_activations = pd.read_csv(network_filename)
```

```python
network_activations["word"] = None
```

```python
for index, row in network_activations.iterrows():
    start_time = row['start_time']
    for index_word, row_word in words.iterrows():
        #print(row['start_time'])
        #print(start_time)
        if row_word['start_time'] > start_time and index_word != 0:
            possible_words = [ words.loc[index_word - 1]['word'], words.loc[index_word]['word']]
            network_activations.at[index, 'word'] = possible_words
            break
            #network_activations[index]['word'] = words[index_word - 1]['word']
```

```python
network_activations.to_csv("network_analysis.csv")
```
